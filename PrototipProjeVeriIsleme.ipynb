```python
import cv2
import os

def resize_images(input_directory, output_directory, target_size):
    # Hedef boyut
    target_width, target_height = target_size

    # Çıkış dizinini oluştur
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Giriş dizinindeki tüm resim dosyalarını al
    image_files = [f for f in os.listdir(input_directory) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]

    for image_file in image_files:
        # Giriş resmini oku
        input_path = os.path.join(input_directory, image_file)
        image = cv2.imread(input_path)

        # Resmi hedef boyuta yeniden boyutlandır
        resized_image = cv2.resize(image, (target_width, target_height))

        # Çıkış resmini kaydet
        output_path = os.path.join(output_directory, image_file)
        cv2.imwrite(output_path, resized_image)

# Örnek kullanım
output_directory = '/content/gdrive/MyDrive/protatip proje/veri/train/Kırık2'
target_size = (224, 224)  # Hedef boyut
resize_images(directory, output_directory, target_size)
```

```python
import cv2
import numpy as np
import os

def normalize_images(input_directory, output_directory):
    # Çıkış dizinini oluştur
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Giriş dizinindeki tüm resim dosyalarını al
    image_files = [f for f in os.listdir(input_directory) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]

    for image_file in image_files:
        # Giriş resmini oku
        input_path = os.path.join(input_directory, image_file)
        image = cv2.imread(input_path)

        # Görüntü piksel değerlerini [0, 1] aralığına ölçekle
        normalized_image = image / 255.0

        # Çıkış resmini kaydet
        output_path = os.path.join(output_directory, image_file)
        cv2.imwrite(output_path, (normalized_image * 255).astype(np.uint8))
```

```python
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt

def apply_augmentation(input_directory, output_directory):
    # Çıkış dizinini oluştur
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Giriş dizinindeki tüm resim dosyalarını al
    image_files = [f for f in os.listdir(input_directory) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]

    for image_file in image_files:
        try:
            # Giriş resmini oku
            input_path = os.path.join(input_directory, image_file)
            image = cv2.imread(input_path)

            # Görüntü döndürme (örneğin, 90 derece saat yönünde)
            rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

            # Görüntüyü yatay simetri ile çoğaltma
            flipped_image = cv2.flip(image, 1)

            #Blur
            blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

            #Öteleme
            rows, cols, _ = image.shape
            translation_matrix = np.float32([[1, 0, 50], [0, 1, 30]])  # X ve Y ekseninde 50 ve 30 birim öteleme
            translated_image = cv2.warpAffine(image, translation_matrix, (cols, rows))

            #Kırpma
            crop_start_x, crop_start_y, crop_width, crop_height = 50, 50, cols-100, rows-100
            cropped_image = image[crop_start_y:crop_start_y+crop_height, crop_start_x:crop_start_x+crop_width]

            # Çıkış resimlerini kaydet
            output_path_rotated = os.path.join(output_directory, f"rotated_{image_file}")
            output_path_flipped = os.path.join(output_directory, f"flipped_{image_file}")
            output_path_blurred = os.path.join(output_directory, f"blurred_{image_file}")
            output_path_translated = os.path.join(output_directory, f"translated_{image_file}")
            output_path_cropped = os.path.join(output_directory, f"cropped_{image_file}")

            cv2.imwrite(output_path_rotated, rotated_image)
            cv2.imwrite(output_path_flipped, flipped_image)
            cv2.imwrite(output_path_blurred, blurred_image)
            cv2.imwrite(output_path_translated, translated_image)
            cv2.imwrite(output_path_cropped, cropped_image)


        except Exception as e:
            print(f"Hata oluştu: {e}")

# Örnek kullanım
input_output = [("/content/gdrive/MyDrive/protatip proje/veri/val/Kırık3","/content/gdrive/MyDrive/protatip proje/veri/val/Kırık4"),
                ("/content/gdrive/MyDrive/protatip proje/veri/val/BeyinTümörü3","/content/gdrive/MyDrive/protatip proje/veri/val/BeyinTümörü4"),
                ("/content/gdrive/MyDrive/protatip proje/veri/val/MelomaCiltKanser3","/content/gdrive/MyDrive/protatip proje/veri/val/MelomaCiltKanser4"),
                ("/content/gdrive/MyDrive/protatip proje/veri/val/MemeKanseri3","/content/gdrive/MyDrive/protatip proje/veri/val/MemeKanseri4")]

for input,output in input_output :
  apply_augmentation(input,output)
```

```python
input_output = [("/content/gdrive/MyDrive/protatip proje/veri/train/Kırık3","/content/gdrive/MyDrive/protatip proje/veri/train/Kırık4"),
                ("/content/gdrive/MyDrive/protatip proje/veri/train/BeyinTümörü3","/content/gdrive/MyDrive/protatip proje/veri/train/BeyinTümörü4"),
                ("/content/gdrive/MyDrive/protatip proje/veri/train/Meloma_Cilt_Kanseri3","/content/gdrive/MyDrive/protatip proje/veri/train/MelomaCiltKanser4"),
                ("/content/gdrive/MyDrive/protatip proje/veri/train/Meme_Kanseri3","/content/gdrive/MyDrive/protatip proje/veri/train/MemeKanseri4")]

for input,output in input_output :
  apply_augmentation(input,output)
```
